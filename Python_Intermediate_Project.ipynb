```python
'''
Q10. Please create a new column named total_bedroom_size. If the total bedrooms
     is 10 or less, it should be quoted as small. If the total bedrooms is 11
     or more but less than 1000, it should be medium, otherwise it should be 
     considered large.
'''

# importing the libraries
import pandas as pd

# making data frame from excel file 
data = pd.read_excel('C:\\Users\\ASUS\\Python Jupiter\\\\2198208-housing_(1).xlsx')

size = [] # declaring a list

for i in data['total_bedrooms']:
    if (i <= 10):
        size.append('SMALL')       # append the element to the end of the list
    elif (i == 11) | (i < 1000):
        size.append('MEDIUM')
    elif (i > 1000):
        size.append('LARGE')
    else:                           # For Null Values
        size.append('None')
        
# print(size)

# Using 'tota_bedroom_size' as the column name
# and equating it to the list
data['total_bedroom_size'] = size

data
```
